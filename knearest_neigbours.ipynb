step 1, import lib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

step 2 Create a small synthetic dataset

```python
X, y = make_classification(n_samples=1000,n_informative=2,n_redundant=0,
                          n_features=2,
                          n_classes=3,
                          n_clusters_per_class=1,
                          random_state=42)
```

step 3 split in train and test

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_test
```

train knn model

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
X_train
```

predit on test set

```python
y_pred=knn.predict(X_test)
y_pred
```

step 6 evaluation


```python
print('Confusion matrix')
print(confusion_matrix(y_test,y_pred))
print('/nclassification report')
print(classification_report(y_test,y_pred))
print('accuracy score')
print(accuracy_score(y_test,y_pred))
```

visualize the dataset


```python
plt.figure(figsize=(6,6))
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap='rainbow',marker='o',edgecolors='k',label='train')
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap='rainbow',marker='*',edgecolors='k',label='test')
plt.title('KNN classification(train=circle, test pred =traingles')
plt.xlabel('feature 1')
plt.ylabel('feature 2')
plt.legend()
plt.show()
```

Plot Breakdown

Axes (feature 1 & feature 2)

These are the 2 features generated by make_classification.

Each point’s position is based on these feature values.

Colors = Classes

Red, green, purple clusters = the 3 classes.

KNN tries to learn the decision boundaries between them.

Markers

Circles (o) = training points.

Stars (★) = test set predictions (colored by predicted class).

KNN Predictions

Notice that most stars fall into the correct cluster of their color.

A few stars fall into “border areas” and get misclassified → these are exactly what showed up as errors in the confusion matrix.

🔹 How to interpret

The clusters are well separated, so KNN (k=3) does a good job here.

Where clusters overlap (e.g., red vs purple), that’s where misclassifications occur.

The visualization confirms what the confusion matrix told us: most predictions correct, some mistakes at boundaries.
