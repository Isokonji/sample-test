TRAIN THE DATASET

```python
from sklearn.model_selection import train_test_split
x=df.loc[:,['cgpa']]
y=df.loc[:,['package']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#get slope and intercept
```

```python
#fit
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#get slope and intercept
print(lr.coef_)
print(lr.intercept_)
```

```python
Y=lr.coef_*x+lr.intercept_
print(Y)
```

```python

#make predictions

y_pred=lr.predict(x_test)
y_pred
```

```python
# ypred is the prediction using x test for the packages , so now we will compare the ytest and the y pred
```

```python
print(y_pred,y_test)
```

```python
# use evaluation metrics to check the model
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
R2=r2_score(y_test,y_pred)
RMSE=np.sqrt(mean_squared_error(y_test,y_pred))
MSE=mean_squared_error(y_test,y_pred)
MAE=mean_absolute_error(y_test,y_pred)
adjusted_r2= 1 - (1-R2)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1) #len is the number of samples used
print(f"Mean Squared Error (MSE): {MSE:.4f}")
print(f"Root Mean Squared Error (RMSE): {RMSE:.4f}")
print(f"R Score: {R2:.4f}")
print(f"Adjusted R Score: {adjusted_r2:.4f}")
```

```python
#plot the linear regression  model
plt.scatter(x=x_train,y=y_train)
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.title('Linear Regression')

plt.show()
```
